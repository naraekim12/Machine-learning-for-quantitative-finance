## Hurst exponent

```python
# returns the Hurst Exponent of the time series
def get_hurst_exponent(time_series, max_lag=20):
    lags = range(2, max_lag)
    # variances of the lagged differences
    tau = [np.std(np.subtract(time_series[lag:], time_series[:-lag])) for lag in lags]
    # calculate the slope of the log plot -> the Hurst Exponent
    reg = np.polyfit(np.log(lags), np.log(tau), 1)
    return reg[0]
```

```python
for lag in [10, 20, 100, 300, 500, 1000, 2000, 3000, 4000]:
    hurst_exp = get_hurst_exponent(EURGBP_data["CLOSE"].values, lag)
    print(f" {lag} lags: {hurst_exp:.4f}")
```

## Volatility

```python
EURGBP_data['Log returns'] = np.log(EURGBP_data['CLOSE']/EURGBP_data['CLOSE'].shift())
EURGBP_data['Log returns'].std()
volatility = EURGBP_data['Log returns'].std()*252**.5
str_vol = str(round(volatility, 4)*100)

fig, ax = plt.subplots()
EURGBP_data['Log returns'].hist(ax=ax, bins=500, alpha=0.6, color='b')
plt.xlim(-0.0012, 0.0012) # the range of x-axis
ax.set_xlabel('Log return')
ax.set_ylabel('Freq of log return')
ax.set_title('Volatility: ' + str_vol + '%')
```

## Random walk: Monte Carlo Simulation

EURGBP

```python
returns = EURGBP_data['CLOSE'].pct_change()
volatility = returns.std()

# Running single simulation
T = 252
count = 0
price_list = []
last_price = EURGBP_data['CLOSE'].iloc[-1]

price = last_price * (1 + np.random.normal(0, volatility))
price_list.append(price)

for y in range(T):
    if count == 251:
        break
    price = price_list[count]* (1 + np.random.normal(0, volatility))
    price_list.append(price)
    count += 1

plt.plot(price_list)
plt.show()

# Running multiple simulations

NUM_SIMULATIONS =  1000
df = pd.DataFrame()
last_price_list = []
for x in range(NUM_SIMULATIONS):
    count = 0
    price_list = []
    price = last_price * (1 + np.random.normal(0, volatility))
    price_list.append(price)
    
    for y in range(T):
        if count == 251:
            break
        price = price_list[count]* (1 + np.random.normal(0, volatility))
        price_list.append(price)
        count += 1
        
    df[x] = price_list
    last_price_list.append(price_list[-1])
        
fig = plt.figure()
plt.plot(df)
plt.xlabel('Day')
plt.ylabel('Price')
plt.show()
```

```python
print("Quantile 5%: ",np.percentile(last_price_list,5))
print("Quantile 95%: ",np.percentile(last_price_list,95))

plt.hist(last_price_list,bins=100)
plt.axvline(np.percentile(last_price_list,5), color='r', linestyle='dashed', linewidth=2)
plt.axvline(np.percentile(last_price_list,95), color='r', linestyle='dashed', linewidth=2)
plt.show()
```
