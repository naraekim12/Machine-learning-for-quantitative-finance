```python
import pandas as pd
import seaborn as sns
import numpy as np 
import matplotlib.pyplot as plt
```

```python
#Read in csv file
#convert the first column 'Dates' to a Timestamp. (Before converting Dates to Timestamp, the datatype was 'object')
forex_data1 = pd.read_csv("CSV_DAT_XLSX_EURGBP_M1_2013.csv", parse_dates=[0] ,names=["DateTime", "OPEN", "HIGH", "LOW", "CLOSE", "VOLUME"] ) 
forex_data2 = pd.read_csv("CSV_DAT_XLSX_EURUSD_M1_2013.csv", parse_dates=[0] ,names=["DateTime", "OPEN", "HIGH", "LOW", "CLOSE", "VOLUME"]) 
forex_data3 = pd.read_csv("CSV_DAT_XLSX_GBPUSD_M1_2013.csv", parse_dates=[0] ,names=["DateTime", "OPEN", "HIGH", "LOW", "CLOSE", "VOLUME"]) 

forex_data1
```

```python
#to make sure the Date is sorted by datetime because it is important in time series modelling
forex_data1 = forex_data1.sort_values('DateTime')
forex_data2 = forex_data2.sort_values('DateTime')
forex_data3 = forex_data3.sort_values('DateTime')
```

```python
#find missing minutes
forex_data1 = forex_data1.set_index('DateTime')
missing_times = pd.date_range(start=forex_data1.index[0], end=forex_data1.index[-1], freq='60s').difference(forex_data1.index)
```

```python
#find missing minutes
forex_data2 = forex_data2.set_index('DateTime')
missing_times2 = pd.date_range(start=forex_data2.index[0], end=forex_data2.index[-1], freq='60s').difference(forex_data2.index)
missing_times2
```

```python
#find missing minutes
forex_data3 = forex_data3.set_index('DateTime')
missing_times3 = pd.date_range(start=forex_data3.index[0], end=forex_data3.index[-1], freq='60s').difference(forex_data3.index)
missing_times3
```

```python
#Find missing times and drop them
```

```python
#reset index back to default to use the DateTime as a column
forex_data3 = forex_data3.reset_index()
forex_data2 = forex_data2.reset_index()
forex_data1 = forex_data1.reset_index()
```

```python
#delete all the missing times from forex_data3
```

```python
for missing in missing_times:
    if (missing in forex_data3['DateTime'].values) == True:
        #forex_data3 = forex_data3.drop(missing, inplace=True,axis=0)
        forex_data3.drop(forex_data3.loc[forex_data3['DateTime']==missing].index, inplace=True)

forex_data3
```

```python
for missing in missing_times2:
    if (missing in forex_data3['DateTime'].values) == True:
        #forex_data3 = forex_data3.drop(missing, inplace=True,axis=0)
        forex_data3.drop(forex_data3.loc[forex_data3['DateTime']==missing].index, inplace=True)

forex_data3
```

```python
#delete all the missing times from forex_data2
```

```python
for missing in missing_times:
    if (missing in forex_data2['DateTime'].values) == True:
        forex_data2.drop(forex_data2.loc[forex_data2['DateTime']==missing].index, inplace=True)

forex_data2
```

```python
for missing in missing_times3:
    if (missing in forex_data2['DateTime'].values) == True:
        forex_data2.drop(forex_data2.loc[forex_data2['DateTime']==missing].index, inplace=True)

forex_data2
```

```python
for missing in missing_times2:
    if (missing in forex_data1['DateTime'].values) == True:
        forex_data1.drop(forex_data1.loc[forex_data1['DateTime']==missing].index, inplace=True)

forex_data1
```
